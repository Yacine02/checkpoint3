# 1. Importations

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2. Affichage des informations générales sur nos données

```python
df = pd.read_csv('titanic-passengers.csv', sep=";")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

# 3.Nettoyage

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)#Remplacer les valeus manquantes de la colonne Age par la moyenne des ages
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)#Remplacer les valeus manquantes de la colonne Embarked par le mode
df.info()
```

```python
df.isnull().sum()
```

```python
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)#Remplacer les valeus manquantes de la colonne Cabin par le mode
df.info()
```

```python
df.isnull().sum()
```

# 3.Transformation

```python
#Toutes les valeurs manquantes sont comblées on peut passer à la transformation
df.head()
```

```python
#Remplacer les valeurs catégorielles Yes et No par des valeurs numériques 1 et 0
dictSurvived = {'Yes': 1, 'No' : 0}
df['Survived'].replace(dictSurvived , inplace=True) 
df.head()
```

```python
#Remplacer les valeurs catégorielles Male et Female par des valeurs numériques 1 et 0
dictSex = {'male': 1, 'female': 0}
df['Sex'].replace(dictSex, inplace=True)
df.head()
```

# Exécution de la fonction et analyse

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map( df )
```

```python
#Utilisation la fonction groupby combinée avec la fonction mean () pour afficher la relation entre Pclass et survived 
df[['Pclass', 'Survived']].groupby(['Pclass'],as_index=True).mean()
```

# Suppression des colonnes inutiles

```python
df.drop('Name',axis=1, inplace=True)
```

```python
df.drop('Ticket',axis=1, inplace=True)
```

```python
df.drop('Fare',axis=1, inplace=True)
```

```python
df.drop('Cabin',axis=1, inplace=True)
```

```python
df.drop('Embarked',axis=1, inplace=True)
```

```python
df.drop('PassengerId',axis=1, inplace=True)
```

```python
df.head()
```

```python
#création d'une nouvelle colonne appelée Titre qui contient le titre approprié pour chaque individu 
name
```

```python
titre = []
for i in name:
    a = i.split(",")
    b = a[1].split(".")
    c = b[0].split(" ")
    titre.append(c[1])

df = df.assign(Title = titre)
```

```python
df.head(50)
```

```python
#Corrélation entre le titre et d'autres caractéristiques (par exemple, le sexe, le tarif, l'âge...)
sns.barplot(x = 'Age', y = 'Title', hue = 'Survived', data = df, palette = 'flare', edgecolor = 'w')
```

```python
#regrouper ces titres en plus utiles et les remplacer dans le tableau.
Title_Dictionary = {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                      "Dr":         "Officer",

                    "Rev":        "Officer",

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                   "Lady" :      "Royalty",

                  "the Countess": "Royalty",

                    "Dona":       "Royalty",

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs",

                    "Master" :    "Master"

                    }
```

```python
df["Title"].replace(Title_Dictionary, inplace=True)
```
